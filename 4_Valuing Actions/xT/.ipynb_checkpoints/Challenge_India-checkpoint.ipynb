## Indian Superleague Analysis on xT

```python
import pandas as pd
import json
# plotting
import os
import pathlib
import warnings 
from joblib import load
from mplsoccer import Pitch
from itertools import combinations_with_replacement
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np
import xgboost as xgb
from scipy.stats import binned_statistic_2d
pd.options.mode.chained_assignment = None
warnings.filterwarnings('ignore')
from mplsoccer import Sbopen
import time
```

```python
parser = Sbopen()
matches = parser.match(competition_id = 1238,season_id = 108)
```

```python
matches = matches['match_id'].to_list()
```

```python
df = pd.DataFrame()
for i in matches:
    print(i)
    match = parser.event(i)[0]
    df = pd.concat([df,match])
    time.sleep(1)
    
df.head()
```

```python
df.shape[0]
```

```python
df['x'] = df['x'] * 105 / 120
df['y'] = df.y.apply(lambda x: (80 - x) * 68/80)
df['end_x'] = df['end_x'] * 105 / 120
df["end_y"] = df.end_y.apply(lambda x: (80 - x) * 68/80)
df['Cx'] = 105 - df['x']
df["Cy"] = df.y.apply(lambda x: abs(x - 68/2))
df["Cend_y"] = df.end_y.apply(lambda x: abs(x - 68/2))
```

```python
var = ['x','end_x','Cy','Cend_y']
#combinations
inputs = []

inputs.extend(combinations_with_replacement(var, 1))
#one variable combinations
inputs.extend(combinations_with_replacement(var, 1))
#2 variable combinations
inputs.extend(combinations_with_replacement(var, 2))
#3 variable combinations
inputs.extend(combinations_with_replacement(var, 3))
```

```python
for i in inputs:
    if len(i)>1:
        col= ''
        x=1
        
        for c in i:
            col+=c
            #print(col)
            
            x = x * df[c]
            
        df[col] = x
        var.append(col)
        
df.head()
```

```python
path_model = os.path.join(str(pathlib.Path().resolve()),'model','model_xT.json')
model = xgb.XGBClassifier()
model.load_model(path_model)
```

```python
mov = df[(df['end_x'].isna()==False) & (df['type_name']!= "Goal Keeper")]
mov.shape[0]
```

```python
mov.type_name.value_counts()
```

```python
mov['possession_id'] = mov.possession.astype(str) + '_' + mov.match_id.astype(str)
len(mov['possession_id'].unique())
```

```python
mov['shot_end'] = 0
mov['xG'] = 0
for i in mov.possession_id.unique():
    k = mov[mov.possession_id==i]
    if k.iloc[-1]["type_name"] == "Shot":
        mov.loc[mov["possession_id"] == i, "shot_end"] = 1
        mov.loc[mov["possession_id"] == i, "xG"] = mov[mov.possession_id==i].iloc[-1]['shot_statsbomb_xg']
```

```python
mov.shot_end.value_counts()
```

```python
passes = mov[mov.type_name!='Shot']
X = passes[var].values
y = passes["shot_end"].values
```

```python
y_pred_proba = model.predict_proba(X)[::,1]
y_pred_proba
```

```python
passes['shot_prob'] = y_pred_proba
shot_ended = passes[passes["shot_end"] == 1]
X_train = shot_ended[var].values
y_train = shot_ended["xG"].values

lr = LinearRegression()
lr.fit(X_train,y_train)

y_pred = lr.predict(X)
passes["xG_pred"] = y_pred

passes['xT'] = passes["xG_pred"] * passes['shot_prob']
passes.xT.isna().any()
```

```python
passes.type_name.value_counts()
```

```python
gr1 = passes.groupby("player_name",as_index=False)['xT'].sum().sort_values(by='xT',ascending=False)
gr1.head(25)
```

```python
gr2 = passes.groupby("player_name",as_index=False).agg({'xT': ['mean', 'sum', 'count']})
cols = []
for i in gr2.columns:
    if i[1]=='':
        cols.append(i[0])
    else:
        cols.append(i[1])
        
gr2.columns = cols
gr2 = gr2.sort_values('mean',ascending=False)
gr2 = gr2[gr2['count']>gr2['count'].quantile(0.7)]
gr2.head(25)
```
