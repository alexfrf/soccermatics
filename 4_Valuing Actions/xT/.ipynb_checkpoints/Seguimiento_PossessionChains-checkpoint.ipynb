## Possession Chains

```python
import pandas as pd
import numpy as np
import json
# plotting
import os
import pathlib
import warnings
import statsmodels.api as sm
import statsmodels.formula.api as smf
from mplsoccer import Pitch
import matplotlib.pyplot as plt

pd.options.mode.chained_assignment = None
warnings.filterwarnings('ignore')
```

```python
file_name = 'events_England.json'
path = os.path.join(str(pathlib.Path().resolve()), 'data', 'Wyscout','events', file_name)
with open(path) as f:
    data = json.load(f)
df = pd.DataFrame(data)
df = df.reset_index()
df.head()
```

### Preparación de los datos

Tratamos de detectar en los datos todas las acciones que impliquen el final de una posesión

```python
df.eventName.value_counts()
```

```python
next_event = df.shift(-1, fill_value=0)
df["nextEvent"] = next_event["subEventName"]
df["kickedOut"] = df.apply(lambda x: 1 if x.nextEvent == "Ball out of the field" else 0, axis = 1)

interruption = df[df['eventName']=='Interruption']
lost_duels = df.loc[df["eventName"] == "Duel"]
df = df.drop(lost_duels[lost_duels.apply (lambda x:{'id':1802} in x.tags, axis = 1)].index)
df = df.drop(interruption.index)
df = df.drop(df[df['subEventName']=='Ball out of the field'].index)
df = df.drop(df[df["subEventName"].isin(["Goalkeeper leaving line", "Save attempt", "Reflexes"])].index)
df.head()
```

### Aislando cada posesión

```python
def isolateChains(df):
    
    df["nextTeamId"] = df.shift(-1, fill_value=0)["teamId"]
    chain_team = df.iloc[0]["teamId"]
    period = df.iloc[0]["matchPeriod"]
    stop_criterion = 0
    chain = 0
    moment = 0
    df["possesion_chain"] = 0
    df["possesion_chain_team"] = 0
    df['possesion_chain_moment'] =0
    
    for i,j in df.iterrows():
        moment+=1
        df.at[i, "possesion_chain"] = chain
        df.at[i, "possesion_chain_team"] = chain_team
        df.at[i, "possesion_chain_moment"] = moment
        # if pass not accurate/lost duel, add 1 to stop criterion
        if j['eventName']=='Pass' or j['eventName']=='Duel':
            if j["teamId"] == chain_team and {"id": 1802} in j["tags"]:
                    stop_criterion += 1
            if j["teamId"] != chain_team and {"id": 1801} in j["tags"]:
                    stop_criterion += 1
        #if ball intercepted properly add 2
        if j["eventName"] == "Others on the ball":
               if j["teamId"] == j["nextTeamId"]:
                    stop_criterion += 2
                    
        #if shot, add 2 to stop criteriom
        if j["eventName"] in ["Shot", "Foul", "Offside"]:
            stop_criterion += 2
        if j["kickedOut"] == 1:
            stop_criterion += 2
            
        #criterion for stopping when half ended    
        if j['matchPeriod'] != period:
            moment = 0
            stop_criterion = 0
            chain += 1
            chain_team = j['teamId']
            period = j["matchPeriod"]
            df.at[i, "possesion_chain"] = chain
            df.at[i, "possesion_chain_moment"] = moment
            df.at[i, "possesion_chain_team"] = chain_team
            
         #possesion chain ended
        if stop_criterion >= 2:
            moment = 0
            chain += 1
            stop_criterion = 0
            chain_team = j['nextTeamId']
            
    return df

df = isolateChains(df)

df.head()
```

```python
# Applying the xg model to the dataframe

def calulatexG(df):
    """
    Parameters
    ----------
    df : dataframe
        dataframe with Wyscout event data.

    Returns
    -------
    xG_sum: dataframe
        dataframe with xG for each shot

    """
    #very basic xG model based on
    shots = df.loc[df["eventName"] == "Shot"].copy()
    shots["X"] = shots.positions.apply(lambda cell: (100 - cell[0]['x']) * 105/100)
    shots["Y"] = shots.positions.apply(lambda cell: cell[0]['y'] * 68/100)
    shots["C"] = shots.positions.apply(lambda cell: abs(cell[0]['y'] - 50) * 68/100)
    #calculate distance and angle
    shots["Distance"] = np.sqrt(shots["X"]**2 + shots["C"]**2)
    shots["Angle"] = np.where(np.arctan(7.32 * shots["X"] / (shots["X"]**2 + shots["C"]**2 - (7.32/2)**2)) > 0, np.arctan(7.32 * shots["X"] /(shots["X"]**2 + shots["C"]**2 - (7.32/2)**2)), np.arctan(7.32 * shots["X"] /(shots["X"]**2 + shots["C"]**2 - (7.32/2)**2)) + np.pi)
    #if you ever encounter problems (like you have seen that model treats 0 as 1 and 1 as 0) while modelling - change the dependant variable to object
    shots["Goal"] = shots.tags.apply(lambda x: 1 if {'id':101} in x else 0).astype(object)
        #headers have id = 403
    headers = shots.loc[shots.apply (lambda x:{'id':403} in x.tags, axis = 1)]
    non_headers = shots.drop(headers.index)

    headers_model = smf.glm(formula="Goal ~ Distance + Angle" , data=headers,
                               family=sm.families.Binomial()).fit()
    #non-headers
    nonheaders_model = smf.glm(formula="Goal ~ Distance + Angle" , data=non_headers,
                               family=sm.families.Binomial()).fit()
    #assigning xG
    df["xG"] = 0
    b_head = headers_model.params
    xG = 1/(1+np.exp(b_head[0]+b_head[1]*headers['Distance'] + b_head[2]*headers['Angle']))
    headers = headers.assign(xG = xG)
    for index, row in headers.iterrows():
        df.at[index, "xG"] = row["xG"]
    #non-headers
    b_nhead = nonheaders_model.params
    xG = 1/(1+np.exp(b_nhead[0]+b_nhead[1]*non_headers['Distance'] + b_nhead[2]*non_headers['Angle']))
    non_headers = non_headers.assign(xG = xG)
    for index, row in non_headers.iterrows():
        df.at[index, "xG"] = row["xG"]

    penalties = df.loc[df["subEventName"] == "Penalty"]
    #treating penalties like shots
    penalties["X"] = 11
    #calculate distance and angle
    penalties["Distance"] = 11
    penalties["Angle"] = np.arctan(7.32 * 11 /(11**2 - (7.32/2)**2))
    #if you ever encounter problems (like you have seen that model treats 0 as 1 and 1 as 0) while modelling - change the dependant variable to object
    penalties["Goal"] = penalties.tags.apply(lambda x: 1 if {'id':101} in x else 0).astype(object)
    penalties = penalties.assign(xG = xG)
    for index, row in penalties.iterrows():
        df.at[index, "xG"] = row["xG"]
    return df

df = calulatexG(df)
```

Tras añadir los xG, el siguiente paso es encontrar aquellas cadenas que acaben con disparo. Si una cadena acaba en falta y posteriormente la jugada acaba en gol, debemos trasladar los xG no sólo a lo que sucede tras la falta, sino también a las acciones que provocaron la misma.

```python
def prepareChains(df):
    df["shot_end"] = 0
    no_chains = max(df["possesion_chain"].unique())
    indicies = []
    
    for i in range(no_chains+1):
        possesion_chain_df = df.loc[df["possesion_chain"] == i]
        #check if the possesion chain is not empty
        if len(possesion_chain_df) > 0:
            if possesion_chain_df.iloc[-1]["eventName"] == "Shot":
                xG = possesion_chain_df.iloc[-1]["xG"]
                df.loc[df["possesion_chain"] == i, "xG"] = xG
                df.loc[df["possesion_chain"] == i, "shot_end"] = 1
                
                p = i-1
                if p > 0:
                    try:
                        prev = df.loc[df["possesion_chain"] == p]
                        while prev.iloc[-1]['eventName']=='Foul':
                            df.loc[df["possesion_chain"] == p, "xG"] = xG
                            df.loc[df["possesion_chain"] == p, "shot_end"] = 1
                            p = p-1
                            prev = df.loc[df["possesion_chain"] == p]
                    except:
                        p = p-1
            indicies_team = possesion_chain_df[possesion_chain_df["teamId"] == possesion_chain_df.teamId.mode().iloc[0]].index.values.tolist()
            indicies.extend(indicies_team)
    df = df.loc[indicies]
    return df

df = prepareChains(df)
df.head()
```

### Preparación de datos para visualizaciones

```python
df= df[df.apply(lambda x: len(x.positions)>=2,axis=1)] # tomamos solo acciones que tengan x,y y su final

df["x"] = df.positions.apply(lambda x: (x[0]['x']) * 105/100)
df['y'] = df.positions.apply(lambda x: (100 - x[0]['y']) * 68/100)
df["Cy"] = df.positions.apply(lambda x: abs(x[0]['y'] - 50) * 68/100)
df["end_x"] = df.positions.apply(lambda x: (x[1]['x']) * 105/100)
df["end_y"] = df.positions.apply(lambda x: (100 - x[1]['y']) * 68/100)
df["Cend_y"] = df.positions.apply(lambda x: abs(x[1]['y'] - 50) * 68/100)
df["Cx"] = df.positions.apply(lambda x: (100 - x[0]['x']) * 105/100)
df["Cend_x"] = df.positions.apply(lambda x: (100 - x[1]['x']) * 105/100)
df.loc[df["eventName"] == "Shot", "end_x"] = 105
df.loc[df["eventName"] == "Shot", "Cend_x"] = 0
df.loc[df["eventName"] == "Shot", "end_y"] = 34
df.loc[df["eventName"] == "Shot", "Cend_y"] = 0
```

```python
#plot possesion chain that ended with shot
chain = df.loc[df["possesion_chain"] == 4]
#get passes
passes = chain.loc[chain["eventName"].isin(["Pass","Free Kick"])]
#get events different than pass
not_pass = chain.loc[~chain["eventName"].isin(["Pass","Free Kick"])].iloc[:-1]
#shot is the last event of the chain (or should be)
shot = chain.iloc[-1]
start = chain.iloc[0]
#plot
pitch = Pitch(line_color='black',pitch_type='uefa', line_zorder = 2)
fig, ax = pitch.grid(grid_height=0.9, title_height=0.06, axis=False,
                     endnote_height=0.04, title_space=0, endnote_space=0)
#passes
pitch.arrows(passes.x, passes.y,
            passes.end_x, passes.end_y, color = "blue", ax=ax['pitch'], zorder =  3)
#shot
pitch.arrows(shot.x, shot.y,
            shot.end_x, shot.end_y, color = "red", ax=ax['pitch'], zorder =  3)
#other passes like arrows
pitch.scatter(start.x, start.y, marker = '*', color = "red", ax=ax['pitch'], zorder =  3, s= 200)
pitch.lines(not_pass.x, not_pass.y, not_pass.end_x, not_pass.end_y, color = "grey", lw = 1.5, ls = 'dotted', ax=ax['pitch'])
ax['title'].text(0.5, 0.5, 'Passes leading to a shot', ha='center', va='center', fontsize=30)
plt.show()
```
